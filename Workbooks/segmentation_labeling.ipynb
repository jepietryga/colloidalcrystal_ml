# Notebook for Segmentation Labeling
This guided notebook should help use the built in labeling feature of the class. This enables one to more easily access the tool

```python
# Imports
import sys
sys.path.append("..")
import pathlib
import os

from facet_ml.segmentation.segmenter import ImageSegmenter
from facet_ml.segmentation.features import facet_score
```

Specify the image of interest as img_oi. The file's name will be the csv file saved in the specified result_folder_path.
Additional file_strings can be added with the "file_str" variable for easy access

```python
img_oi = "/Users/jacobpietryga/Desktop/Academics/colloidal_crystal_ML/Images/Training/4 nM 3.bmp"
IS = ImageSegmenter(img_oi,
                    edge_modification="localthresh",
                    threshold_mode="ensemble",
                    result_folder_path="../Results",
                    file_str="test"
                    )
facet_score(IS)
IS.create_csv()
```

```python
IS.begin_labeling()
```
